```python
import pandas as pd
import numpy as np
from StringIO import StringIO
from zipfile import ZipFile
from urllib import urlopen
import requests
import csv
import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline
```

```python
def unzip_file(year):
    file_path = 'data/{}_sqf_csv.zip'.format(year)
    zipdir = ZipFile(file_path)
    zipdir.extractall('data')
```

```python
for year in range(2003,2016):
    unzip_file(year)
```

```python
def read_file_count(year):
    file_path = 'data/{}.csv'.format(year)
    with open(file_path, 'r') as csv_file:
        reader = csv.reader(csv_file,delimiter = ",")
        count = 0
        for row in reader:
            count += 1
        return count
```

```python
df = {}
for year in range(2003, 2016):
    df['{}'.format(year)] = read_file_count(year)
```

```python
df
```

```python
df_sqf = pd.DataFrame.from_dict(df, orient='index')
```

```python
df_sqf.columns = ['Count']
```

```python
df_sqf = df_sqf.sort()
```

```python
sns.set(color_codes=True)
plt.figure(figsize=(15,8))
plt.plot(df_sqf)
plt.title('Count of SQF Incidents Recorded Over Time')
```
